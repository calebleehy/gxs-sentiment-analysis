```python
df_2['date'] = pd.to_datetime(df_2['date'], format="%d/%m/%Y %H:%M")

df_2['date'] = df_2['date'].dt.date

df_2.head()
```

```python
df_3 = df_2.rename(columns={"month_year": "date"})
df_3.head()
```

```python
df_3["id"] = df_3.index + 1
df_3.head()
```

```python
import random
```

```python
fake_values_A = ['A1', 'A2', 'A3']
fake_values_B = ['B1', 'B2', 'B3']
fake_values_C = ['C1', 'C2', 'C3']
fake_values_D = ['D1', 'D2', 'D3']

def assign_value(row):
    if row['service'] == 'A':  # Example condition: if col1 is even
        return random.choice(fake_values_A)
    elif row['service'] == 'B':
        return random.choice(fake_values_B)
    elif row['service'] == 'C':
        return random.choice(fake_values_C)
    else:
        return random.choice(fake_values_D)
```

```python
df_3['issue'] = df_3.apply(assign_value, axis=1)
df_3.head()
```

```python
df_4 = df_3[['id', 'date', 'bank', 'client', 'rating', 'sentiment', 'intent', 'service','issue', 'review']]
df_4.head()
```

```python
df_4.info()
```

```python
df_4.to_csv("placeholder_data.csv", index = False)
```
