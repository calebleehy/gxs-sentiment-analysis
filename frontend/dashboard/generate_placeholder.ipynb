```python
import pandas as pd
df = pd.read_csv("data.csv")
df.head()
```

```python
df_2 = df.loc[:, ~df.columns.isin(['developer_response', 'recommendation'])]
df_2.head()
```

```python
df_2.dtypes
```

```python
df_2['date'] = df_2['date'].astype("string")
```

```python
df_2.dtypes
```

```python
from datetime import datetime
```

```python
df_2['date'].head()
```

```python
df_2['date'] = df_2['date'].apply(lambda x: datetime.strptime(x, "%d/%m/%Y %H:%M"))
df_2['month_year'] = df_2['date'].apply(lambda x: x.strftime("%m-%y"))
df_2 = df_2.loc[:, ~df_2.columns.isin(['date'])]
```

```python
df_3 = df_2.rename(columns={"month_year": "date"})
df_3.head()
```

```python
df_3["id"] = df_3.index + 1
df_3.head()
```

```python
df_4 = df_3[['id', 'date', 'bank', 'client', 'rating', 'sentiment', 'intent', 'service','review']]
df_4.head()
```

```python
df_4.info()
```

```python
df_4.to_csv("placeholder_data.csv")
```
